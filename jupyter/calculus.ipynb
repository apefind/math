```python
import math
import numpy as np
from sympy import *
from sympy import symbols, integrate
from sympy.utilities.lambdify import lambdify
from sympy.abc import x, y, pi

init_printing()
_out("all")
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plot
import numpy as np
from pylab import meshgrid


def plot_mapping(f, a=-1.0, b=1.0, s=0.05, size=(8, 6)):
    X, Y = meshgrid(np.arange(a, b, s), np.arange(a, b, s))
    Z = f(X, Y)
    fig = plot.figure()
    fig.set_size_inches(*size)
    ax = fig.gca(projection='3d')
    ax.plot_surface(X,
                    Y,
                    Z,
                    cstride=1,
                    rstride=1,
                    linewidth=0,
                    antialiased=False)
    plot.show()
```

```python
def f(x, y):
    return x * y * (x**2 - y**2) / (x**2 + y**2)


u = f(x, y)
```

#### Multivariable mapping

```python
f(x, y)
```

```python
a, b, s = -10.0001, 10.0, 0.1  # ommitting the origin (0, 0)
plot_mapping(f, a=a, b=b, s=s)
```

#### Partial derivatives

```python
simplify(u.diff(x))
simplify(u.diff(y))
```

```python
plot_mapping(lambdify((x, y), u.diff(x), modules='numpy'), a=a, b=b, s=s)
```

```python
simplify(u.diff(x).diff(y))
simplify(u.diff(y).diff(x))
```

#### Mixed partial derivatives

```python
plot_mapping(lambdify((x, y), u.diff(y).diff(x), modules='numpy'),
             a=a,
             b=b,
             s=s)
```

### Fourier Analysis

#### Function Plots

```python
def plot_periodic_function(f, s=0.05, size=(15, 2)):
    X = np.arange(-np.pi, np.pi, s)
    Y = f(X)
    fig, ax = plot.subplots()
    fig.set_size_inches(*size)
    ax.plot(X, Y)
    ax.grid()
    plot.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
    plot.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])
    plot.show()
```

```python
def a_sin_n(x, a=1.0, n=1):
    return a * np.sin(n * x)


def b_cos_m(x, b=1.0, m=1):
    return b * np.cos(m * x)
```

```python
a = 1.0
for n in range(1, 4):
    plot_periodic_function(lambda x: a_sin_n(x, a, n))
```

```python
b = 0.5
for m in range(1, 4):
    plot_periodic_function(lambda x: b_cos_m(x, b, m))
```

```python
a = b = 1.0
a, b = 0.3, 0.1
a, b = 0.9, 0.3
for n in range(1, 4):
    plot_periodic_function(lambda x: a_sin_n(x, a, n) + b_cos_m(x, b, n))
```

#### Differentiation and Integration

```python
t = symbols('t')
n = symbols('n')

def f(x, t):
    return sin(n * x) * cos(n * t)

u = f(x, t)
```

```python
simplify(u.diff(x))
simplify(u.diff(t))
```

```python
u.diff(x).diff(x)
u.diff(t).diff(t)
```

```python
def f(x):
    n = symbols('n')
    return sin(n * x) * sin(n * x)

f(x)
```

```python
integrate(f(x), x)
simplify(integrate(f(x), x))
```

```python
def f(x):
    m, n = symbols('m n')
    return cos(m * x) * sin(n * x)

f(x)
```

```python
integrate(f(x), x)
```

```python
integrate(f(x), (x, -pi, pi))
```

```python
def g(x):
    m, n = symbols('m n')
    return cos(m * x) * cos(n * x)

g(x)
```

```python
integrate(g(x), x)
```

```python
integrate(g(x), (x, -pi, pi))
```
