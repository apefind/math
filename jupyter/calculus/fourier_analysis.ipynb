#### Function Plots

```python
def plot_periodic_function(f, s=0.05, color=None, linewidth=None, size=(15, 2), filepath=None):
    X = np.arange(-np.pi, np.pi, s)
    Y = f(X)
    fig, ax = plot.subplots()
    fig.set_size_inches(*size)
    ax.plot(X, Y, color=color, linewidth=linewidth)
    ax.grid()
    plot.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
    plot.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])
    plot.show()
    if filepath is not None:
        fig.savefig(filepath, dpi=fig.dpi)
```

```python
def a_sin_n(x, a=1.0, n=1):
    return a * np.sin(n * x)


def b_cos_m(x, b=1.0, m=1):
    return b * np.cos(m * x)
```

```python
a = 1.0
for n in range(1, 4):
    plot_periodic_function(lambda x: a_sin_n(x, a, n), linewidth=0.5)
```

```python
b = 0.5
for m in range(1, 4):
    plot_periodic_function(lambda x: b_cos_m(x, b, m), linewidth=0.5)
```

```python
a = b = 1.0
a, b = 0.3, 0.1
a, b = 0.9, 0.3
for n in range(1, 4):
    plot_periodic_function(lambda x: a_sin_n(x, a, n) + b_cos_m(x, b, n), linewidth=0.5)
```

#### Differentiation and Integration

```python
t = symbols('t')
n = symbols('n')

def f(x, t):
    return sin(n * x) * cos(n * t)

u = f(x, t)
```

```python
simplify(u.diff(x))
simplify(u.diff(t))
```

```python
u.diff(x).diff(x)
u.diff(t).diff(t)
```

```python
def f(x):
    n = symbols('n')
    return sin(n * x) * sin(n * x)

f(x)
```

```python
simplify(integrate(f(x), x))
```

```python
def f(x):
    m, n = symbols('m n')
    return cos(m * x) * sin(n * x)

f(x)
```

```python
integrate(f(x), x)
```

```python
integrate(f(x), (x, -pi, pi))
```

```python
def g(x):
    m, n = symbols('m n')
    return cos(m * x) * cos(n * x)

g(x)
```

```python
integrate(g(x), x)
```

```python
integrate(g(x), (x, -pi, pi))
```

#### Approximation absolute value function

```python
def f(x):
    m = symbols('m', integer=True)
    return x * cos(m * x)

f(x)
integrate(f(x), x)
integrate(f(x), (x, 0, pi))
```

```python
def s_k(x, k=1):
    m = (2 * k + 1)
    return  - 4 / np.pi * np.cos(m * x) / m ** 2
```

```python
for k in range(4):
    plot_periodic_function(lambda x: s_k(x, k), linewidth=0.5)
```

```python
def S_n(x, n=4):
    y = 0.0
    for k in range(n):
        m = (2 * k + 1)
        y += np.cos(m * x) / m ** 2
    return np.pi / 2 - 4 / np.pi * y
```

```python
for n in [2, 4, 16]:
    filepath = f'plots/abs_value_n={n}.png'
    plot_periodic_function(lambda x: S_n(x, n), filepath=filepath)
```
